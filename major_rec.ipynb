```python
import pandas as pd 
```

```python
df = pd.read_excel('Higher Education Students Survey (Responses) - Form responses 1.xlsx')
df.columns = ['timestamp', 'email', 'gender', 'age', 'highest_degree', 'area_of_study', 'bachelor_major', 'master_major', 'further_study_yn', 'further_study_major_yn', 'further_study_major', 'short_learning_yn', 'short_learning_courses', 'top3_criteria']
df.head()
```

```python
df['top3_criteria'] = df['top3_criteria'].apply(lambda x: x.split(', '))
```

```python
df_criteria = pd.get_dummies(df['top3_criteria'].apply(pd.Series).stack()).sum(level=0)
df_criteria
```

```python
df = pd.concat([df, df_criteria], axis=1)
```

```python
df.drop(['timestamp', 'email', 'top3_criteria'], axis=1, inplace=True)
```

```python
df.head(5)
```

```python
df_obj = df.select_dtypes(['object'])
df[df_obj.columns] = df_obj.apply(lambda x: x.str.strip())
```

```python
df['bachelor_major']
```

```python
df = pd.get_dummies(data=df, columns=['gender', 'age', 'highest_degree', 'area_of_study', 'bachelor_major', 'master_major', 'further_study_yn', 'further_study_major_yn', 'short_learning_yn', 'short_learning_courses'])
df
```

```python
df = df[~df['further_study_major'].isnull()].reset_index(drop=True)
df
```

```python
df.columns.tolist()
```

```python
import re
df = df.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_digits
# from sklearn.metrics import plot_confusion_matrix 


x = df.drop(['further_study_major'],axis=1)
y = df.further_study_major
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 21, test_size = 0.25) 
```

```python
x_train
```

```python
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
# import lightgbm as lgb


from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(y_train)


# model = lgb.LGBMClassifier()
# model = XGBClassifier(random_state=21).fit(x_train,y_train)
model = MLPClassifier(hidden_layer_sizes=(100,50,25), activation='relu', solver='adam', max_iter=500).fit(x_train, y_train)
# model = RandomForestClassifier(random_state=21).fit(x_train,y_train)


model.fit(x_train, y_train)

y_pred = model.predict(x_test)
y_pred = le.inverse_transform(y_pred)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
report = pd.DataFrame(classification_report(y_test, y_pred, output_dict=True)).transpose()
report
```

```python
y_test1 = pd.DataFrame(y_test)
y_test1['pred'] = y_pred
y_test1 =  y_test1.reset_index()
y_test1.to_csv('mlp_y_test_pred.csv')
```
