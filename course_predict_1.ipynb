```python
df = df.drop(['student_gender', 'student_age', 'student_degree', 'student_name', 'student_email'], axis=1)
df.head(1)
```

```python
import ast

df.course_id = df.course_id.apply(ast.literal_eval)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

s = df['course_id']

mlb = MultiLabelBinarizer()

course_dummies = pd.DataFrame(mlb.fit_transform(s),columns=mlb.classes_, index=df.index)
```

```python
df_basic = df[['id', 'course_id', 'enrolled_course', 'student_course']]
```

```python
df_new = pd.concat([df_basic, course_dummies], axis=1)
```

```python
df_new = pd.get_dummies(df_new, columns=['student_course'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_digits
from sklearn.metrics import ConfusionMatrixDisplay 


x = df_new.drop(['enrolled_course'],axis=1)
y = df_new.enrolled_course
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 21, test_size = 0.25) 
```

```python
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(y_train)



model = XGBClassifier().fit(x_train.drop(['course_id', 'id'],axis=1),y_train)

# model = MLPClassifier(hidden_layer_sizes=(100,50,25), activation='relu', solver='adam', max_iter=500).fit(x_train.drop(['course_id', 'id'],axis=1),y_train)

y_pred = model.predict(x_test.drop(['course_id', 'id'],axis=1))
y_pred = le.inverse_transform(y_pred)
```

```python
y_pred_proba = model.predict_proba(x_test.drop(['course_id', 'id'],axis=1))
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
_classes = []
for k in range(2500):
    _classes.append(model.classes_.tolist())


pred_dict = {}

for idx, item in enumerate(y_pred_proba):
    pred_dict[idx] = [model.classes_.tolist(), item]
```

```python
pred_df = pd.DataFrame(pred_dict)

# pred_df.rename(index={0: 'classes', 1: 'pred_proba'}, inplace=True)




pred_df = pred_df.T


pred_df.columns = ['classes', 'pred_proba']



pred_df['num_pred'] = [len(a) for a in pred_df['pred_proba']]

pred_df['course_id'] = x_test['course_id'].values

pred_df['id'] = x_test['id'].values



_full_candidate_pred = []
_full_max_pred_class = []


for index, row in pred_df.iterrows():
    _candidate_pred = []

    for item in row['course_id']:
        for cls_idx, cls in enumerate(row['classes']):
            if item == cls:
                _candidate_pred.append(row['pred_proba'][cls_idx])
    
    _full_candidate_pred.append(_candidate_pred)


    _full_max_pred_class.append(row['course_id'][_candidate_pred.index(max(_candidate_pred))])





pred_df['candidate_pred'] = _full_candidate_pred
pred_df['recommend_course'] = _full_max_pred_class

# pred_df['actual_course'] = y_test.values

# pred_df.to_csv('pred_result_mlp_student_course.csv')
```
