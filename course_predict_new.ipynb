```python
import pandas as pd

# course = pd.read_csv('e:\Work\huda\course.csv', low_memory=False, index_col='course_id')
course = pd.read_csv('/Users/porea/Library/CloudStorage/OneDrive-UNSW/Backup/Work1/huda/course.csv', low_memory=False, index_col='course_id')
course = course.drop_duplicates()

course = course[~course.index.duplicated(keep='first')]

course.head(10)
```

```python
# course.to_csv('course_new.csv')
```

```python
# df = pd.read_csv('E:\Work\huda\output.csv', low_memory=False)
df = pd.read_csv('/Users/porea/Library/CloudStorage/OneDrive-UNSW/Backup/Work1/huda/output.csv', low_memory=False)

df.head(10)
```

```python
from ast import literal_eval
df.course_id = df.course_id.apply(literal_eval)
```

```python
df_matrix = pd.DataFrame()

group_idx = 0

for index, row in df.iterrows():
    for candidate in row['course_id']:

        _df = pd.DataFrame(course.loc[candidate]).T.reset_index().rename(columns={'index': 'course_id'})
        _df['id'] = row['id']
        _df['student_gender'] = row['student_gender']
        _df['student_age'] = row['student_age']
        _df['student_course'] = row['student_course']
        _df['student_degree'] = row['student_degree']
        if row['enrolled_course'] == candidate:
            _df['is_enrolled'] = 1
        else:
            _df['is_enrolled'] = 0

        _df['group_idx'] = group_idx

        df_matrix = pd.concat([df_matrix, _df]).reset_index(drop=True)
    group_idx += 1
```

```python
df_matrix.head(20)
```

```python
df_matrix['course_cost'] = df_matrix['course_cost'].apply(lambda x: x.replace('K', '000'))
```

```python
df_matrix['course_cost'] = df_matrix['course_cost'].astype('int')
df_matrix['uni_rank'] = df_matrix['uni_rank'].astype('int')
```

```python
df_matrix
```

```python
# xgb_pairwise

# xgb_ndcg

# lightgbm
```

```python
target = 'full'
target_model = 'xgb_pairwise'
```

```python
# keep uni_rank
if target == '1':
    df_matrix_trim = df_matrix.loc[:, ~df_matrix.columns.isin(['course_cost','course_language','course_duration', 'uni_location'])]
    df_matrix_trim

# keep uni rank and course_cost
if target == '2':
    df_matrix_trim = df_matrix.loc[:, ~df_matrix.columns.isin(['course_language','course_duration', 'uni_location'])]
    df_matrix_trim

# full
if target == 'full':
    df_matrix_trim = df_matrix
```

```python
# keep uni_rank
if target == '1':
    df_new = pd.get_dummies(df_matrix_trim, columns=['student_gender', 'student_course', 'student_degree', 'course_name', 'uni_name'])

# keep uni_rank and course_cost
if target == '2':
    df_new = pd.get_dummies(df_matrix_trim, columns=['student_gender', 'student_course', 'student_degree', 'course_name', 'uni_name', 'course_cost'])

# full
if target == 'full':
    df_new = pd.get_dummies(df_matrix_trim, columns=['student_gender', 'student_course', 'student_degree', 'course_name', 'course_duration', 'uni_name', 'uni_location', 'course_language'])
```

```python
import re
df_new = df_new.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', 'a', x))
```

```python
df_new.columns.duplicated()
```

```python
from sklearn.model_selection import GroupShuffleSplit
gss = GroupShuffleSplit(test_size=.01, 
                        n_splits=1, 
                        random_state = 21).split(df_new, groups=df_new['group_idx'])
_, X_test_inds = next(gss)
```

```python
# fake the test set
gss = GroupShuffleSplit(test_size=.002, 
                        n_splits=1, 
                        random_state = 21).split(df_new, groups=df_new['group_idx'])
X_train_inds, _ = next(gss)

```

```python

train_data = df_new.iloc[X_train_inds]
# train_data = df_new
```

```python
train_data
```

```python
X_train = train_data.loc[:, ~train_data.columns.isin(['course_id','id','is_enrolled'])]
y_train = train_data.loc[:, train_data.columns.isin(['is_enrolled'])]
```

```python
groups = train_data.groupby('id').size().to_frame('size')['size'].to_numpy()
```

```python
test_data= df_new.iloc[X_test_inds]
```

```python
X_test = test_data.loc[:, ~test_data.columns.isin(['course_id','is_enrolled'])]
y_test = test_data.loc[:, test_data.columns.isin(['is_enrolled'])]
```

```python
from xgboost import XGBRanker
import lightgbm

model = XGBRanker(  
    tree_method='hist',
    booster='gbtree',
    objective='rank:pairwise',
    random_state=42, 
    learning_rate=0.1,
    colsample_bytree=0.9, 
    eta=0.05, 
    # max_depth=6, 
    max_depth=1, 
    n_estimators=110, 
    # n_estimators=1, 
    subsample=0.75 
    )

if target_model == 'xgb_pairwise':
    model = XGBRanker(  
        random_state=21,
        objective='rank:pairwise',
        )

if target_model == 'xgb_ndcg':
    model = XGBRanker(  
        random_state=21,
        objective='rank:ndcg',
        )

if target_model == 'lightgbm':
    model = lightgbm.LGBMRanker(
        # objective="lambdarank",
        metric="ndcg",
    )


model.fit(X_train, y_train, group=groups, verbose=True)
```

```python
def predict(model, df):
    return model.predict(df.loc[:, ~df.columns.isin(['id'])])


predictions = (X_test.groupby('id')
               .apply(lambda x: predict(model, x)))
```

```python
import numpy as np

pred_df = pd.DataFrame(predictions, columns=['pred'])
pred_df['max_pred'] = pred_df['pred'].apply(lambda x: max(x))
pred_df['max_pred_idx'] = pred_df['pred'].apply(lambda x: np.argmax(x))
pred_df
```

```python
pred_df['y_test'] = test_data.groupby('id')['is_enrolled'].apply(list)
pred_df['max_y_test'] = pred_df['y_test'].apply(lambda x: max(x))
pred_df['max_y_test_idx'] = pred_df['y_test'].apply(lambda x: np.argmax(x))
pred_df
```

```python
from sklearn.metrics import classification_report
report = pd.DataFrame(classification_report(pred_df['max_y_test_idx'], pred_df['max_pred_idx'], output_dict=True)).transpose()
```
